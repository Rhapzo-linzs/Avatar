```python
while True:
    print("\n=== MENU PROGRAM ===")
    print("1. Konversi Hari ke Tahun, Bulan, Hari")
    print("2. FizzBuzz (1-100)")

    pilihan = input("Pilih menu (1/2): ")

    if pilihan == "1":
        print("\n=== Konversi Hari ===")
        x = int(input("Masukkan jumlah hari: "))

        tahun = x // 365
        sisa = x % 365

        bulan = sisa // 30
        hari = sisa % 30

        print(f"Hasil: {tahun} tahun, {bulan} bulan, {hari} hari")

    elif pilihan == "2":
        print("\n=== FizzBuzz ===")
        for i in range(1, 100):
            if i % 3 == 0 and i % 5 == 0:
                print("fizzbuzz")
            elif i % 3 == 0:
                print("fizz")
            elif i % 5 == 0:
                print("buzz")
            else:
                print(i)

    else:
        print("Pilihan tidak valid, coba lagi!")
```
