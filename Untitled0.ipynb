```python
# Data mahasiswa
nim = ["10121001", "10121002", "10121003"]
nama = ["Reyhan", "Ivan", "Miska"]

# Data nilai (2D array)
nilai = [
    [50, 70, 40, 80],  # reyhan
    [78, 78, 80, 65],  # ivan
    [57, 88, 67, 69]   # miska
]

# Rata-rata tiap mahasiswa
rata_mahasiswa = []

for i in range(len(nilai)):
    total = sum(nilai[i])
    rata = total / len(nilai[i])
    rata_mahasiswa.append(rata)

# Cari mahasiswa terbaik
max_rata = max(rata_mahasiswa)
index_terbaik = rata_mahasiswa.index(max_rata)


# Rata-rata tiap mata kuliah
jumlah_mk = len(nilai[0])
rata_mk = []

for j in range(jumlah_mk):
    total = 0
    for i in range(len(nilai)):
        total += nilai[i][j]
    rata = total / len(nilai)
    rata_mk.append(rata)

# Cari MK dengan nilai terkecil
min_rata = min(rata_mk)
index_mk = rata_mk.index(min_rata)

# Output
print("Mahasiswa Terpintar :", nama[index_terbaik],
      "(Nilai :", round(max_rata, 2), ")")

print("Mata Kuliah Nilai Terkecil : MK" + str(index_mk + 1),
      "(Nilai :", round(min_rata, 2), ")")
```

```python
# =========================
# INPUT DATA MAHASISWA
# =========================
jumlah = int(input("Masukkan jumlah mahasiswa: "))

nama = []
nilai = []

for i in range(jumlah):
    print(f"\nMahasiswa ke-{i+1}")
    n = input("Nama: ")
    nama.append(n)

    mk = []
    for j in range(4):
        val = int(input(f"Nilai MK{j+1}: "))
        mk.append(val)
    nilai.append(mk)

# =========================
# FUNGSI TAMPIL DATA
# =========================
def tampil_data():
    print("\nData Mahasiswa:")
    for i in range(len(nama)):
        print(nama[i], ":", nilai[i])

# =========================
# OPERASI MATRIKS
# =========================
def tambah_matriks(A, B):
    hasil = []
    for i in range(len(A)):
        baris = []
        for j in range(len(A[0])):
            baris.append(A[i][j] + B[i][j])
        hasil.append(baris)
    return hasil

def kurang_matriks(A, B):
    hasil = []
    for i in range(len(A)):
        baris = []
        for j in range(len(A[0])):
            baris.append(A[i][j] - B[i][j])
        hasil.append(baris)
    return hasil

def kali_matriks(A, B):
    hasil = []
    for i in range(len(A)):
        baris = []
        for j in range(len(B[0])):
            total = 0
            for k in range(len(B)):
                total += A[i][k] * B[k][j]
            baris.append(total)
        hasil.append(baris)
    return hasil

# =========================
# MENU
# =========================
while True:
    print("\n=== MENU ===")
    print("1. Tampilkan Data")
    print("2. Penjumlahan Matriks")
    print("3. Pengurangan Matriks")
    print("4. Perkalian Matriks")
    print("5. Exit")

    pilih = input("Pilih menu: ")

    if pilih == "1":
        tampil_data()

    elif pilih == "2":
        print("\nInput Matriks B (ukuran sama)")
        B = []
        for i in range(len(nilai)):
            row = []
            for j in range(len(nilai[0])):
                val = int(input(f"B[{i}][{j}]: "))
                row.append(val)
            B.append(row)

        hasil = tambah_matriks(nilai, B)
        print("Hasil Penjumlahan:", hasil)

    elif pilih == "3":
        print("\nInput Matriks B (ukuran sama)")
        B = []
        for i in range(len(nilai)):
            row = []
            for j in range(len(nilai[0])):
                val = int(input(f"B[{i}][{j}]: "))
                row.append(val)
            B.append(row)

        hasil = kurang_matriks(nilai, B)
        print("Hasil Pengurangan:", hasil)

    elif pilih == "4":
        print("\nInput Matriks B")
        kolom_B = int(input("Jumlah kolom matriks B: "))
        B = []

        for i in range(len(nilai[0])):  # harus sesuai kolom A
            row = []
            for j in range(kolom_B):
                val = int(input(f"B[{i}][{j}]: "))
                row.append(val)
            B.append(row)

        hasil = kali_matriks(nilai, B)
        print("Hasil Perkalian:", hasil)

    elif pilih == "5":
        print("Program selesai.")
        break

    else:
        print("Pilihan tidak valid!")
```
